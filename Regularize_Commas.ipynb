```python
import csv
```

```python
lines_to_write = []
with open("eval_segments(1).csv", "r") as f:
    reader = csv.reader(f, delimiter="\t")
    for i, line in enumerate(reader):
        print(i)
        if i < 3: # header rows
            lines_to_write.append(line[0])
        else: #actual rows
            line_to_write = line
            comma_count = 0
            for j, ch in enumerate(line_to_write[0]):
                if ch == ',':
                    if comma_count < 3:
                        comma_count += 1
                    else:
                        line_to_write[0] = line_to_write[0][:j] + ';' + line_to_write[0][j+1:]
            lines_to_write.append(line_to_write[0])
f.close()

print(lines_to_write)

with open("eval_segments_reg.csv", "w") as fout:
    for ltw in lines_to_write:
        fout.write(ltw + '\n')
fout.close()
```
